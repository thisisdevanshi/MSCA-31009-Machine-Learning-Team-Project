## 4. Dataset preperation

**Pre processing**
- Tokenization
- Lowercasing the words
- Lemmatization/Stemming (shorten words to their root stems)
    - *Stemming usually refers to a crude heuristic process that chops off the ends of words in the hope of achieving this goal correctly most of the time, and often includes the removal of derivational affixes.
    - *Lemmatization usually refers to doing things properly with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word*
- Removing stop words
- Cleaning

**Making all the text lowercase**

```python
#lower case all the text
train_df['text']=train_df.apply(lambda row:row['text'].lower(),axis=1)
test_df['text']=test_df.apply(lambda row:row['text'].lower(),axis=1)
```

**Removing the links from the text**

```python
train_df['text']=train_df.apply(lambda row:re.sub(r"//t.co/\w+","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"//t.co/\w+","",row['text']),axis=1)
```

**Removing the @ and #**

- Have removed the text after @'s as they are usually proper nouns and tweet handles. We could add it later
- Have kept the text after #'s

```python
#for #
train_df['text']=train_df.apply(lambda row:re.sub(r"#","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"#","",row['text']),axis=1)
```

```python
#for @
train_df['text']=train_df.apply(lambda row:re.sub(r"@\w+","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"@\w+","",row['text']),axis=1)
```

## 5. Visualizations

**Class balance**

```python
#function to get distrubution of the cat variables
def get_distr_cat(df,feature):
    sns.set()
    plt.figure(figsize=(10,8))
    sns.barplot(x=df[feature].value_counts().index,y=df[feature].value_counts(),palette="husl")
```

```python
get_distr_cat(train_df,'target')
```

**Understanding the most common tokens used**


 **Word Vector**
- Multi dimensional Representation/ array of a word
- Usually sparse



```python
#lower case all the text
train_df['text']=train_df.apply(lambda row:row['text'].lower(),axis=1)

#tokenize
tokenized_docs=[word_tokenize(doc) for doc in train_df['text']]

dictionary=Dictionary(tokenized_docs)

#most common stuff used in the data
sorted(dictionary.token2id.items(), key=lambda kv: kv[1],reverse=True)
```

**Creating gensim bag of words corpus**

**Topic Identification** - **Bag of words approach**
- Create tokens and count of all the tokens
- The more frequent a token is, the more important that word might be


```python
#using token id and frequency
corpus=[dictionary.doc2bow(doc) for doc in tokenized_docs ]
corpus
```

- This is a list of lists where each list item is a document. The first item in the tuple represents the tokenid and the second item represents token frequency
- Easily saved updated and reused

**TF-IDF Model**

**TF-IDF** : Term Frequency Inverse Document Frequency
- Determine the most important words in each document 
- Idea: Each corpus may have more shared words than just the stopwords. These words need to be removed or down weighted in importance
- It is given by:
<img src ="https://miro.medium.com/max/3604/1*qQgnyPLDIkUmeZKN2_ZWbQ.png">
- The weight would be low if a term doesn't appear alot in the document as tf would be low, however it'll also be low if the log is close to 0 i.e. value in bracket is close to 1. So words that occour across many or all documents would have low TF-IDF weight

```python
tfidf=TfidfModel(corpus)
tfidf[corpus[0]]
```

```python
train_df.iloc[0,:]['text']
```

```python
# Sort the weights from highest to lowest: sorted_tfidf_weights
sorted_tfidf_weights = sorted(tfidf[corpus[0]], key=lambda w: w[1], reverse=True)
sorted_tfidf_weights
```
