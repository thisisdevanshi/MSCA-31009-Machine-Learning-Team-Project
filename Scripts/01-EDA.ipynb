## 4. Dataset preperation

**Shape of the dataframe**

```python
print("The train is {}".format(train_df.shape))
print("The test is {}".format(test_df.shape))
```

**Looking at the null values**

```python
(train_df.isnull().sum()/train_df.shape[0])*100
```

- 33% of the tweets don't have locations in the training dataset

```python
(test_df.isnull().sum()/test_df.shape[0])*100
```

- 33% of the tweets don't have locations in the training dataset

**Making all the text lowercase**

```python
#lower case all the text
train_df['text']=train_df.apply(lambda row:row['text'].lower(),axis=1)
test_df['text']=test_df.apply(lambda row:row['text'].lower(),axis=1)
```

**Removing the links from the text**

```python
train_df['text']=train_df.apply(lambda row:re.sub(r"//t.co/\w+","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"//t.co/\w+","",row['text']),axis=1)
```

**Removing the @ and #**

- Have removed the text after @'s as well as they are usually proper nouns and tweet handles. We could add it later
- Have kept the text after #'s

```python
#for #
train_df['text']=train_df.apply(lambda row:re.sub(r"#","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"#","",row['text']),axis=1)
```

```python
#for @
train_df['text']=train_df.apply(lambda row:re.sub(r"@\w+","",row['text']),axis=1)
test_df['text']=test_df.apply(lambda row:re.sub(r"@\w+","",row['text']),axis=1)
```

## 5. Visualization

**Class balance**

```python
#function to get distrubution of the cat variables
def get_distr_cat(df,feature):
    sns.set()
    plt.figure(figsize=(10,8))
    sns.barplot(x=df[feature].value_counts().index,y=df[feature].value_counts(),palette="husl")
```
